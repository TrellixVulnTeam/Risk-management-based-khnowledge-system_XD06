```python
def format(data,columns):
    for c in columns:
        for r in data.index:
            l = []
            data[c][r] = data[c][r].strip()
            if ' ' in data[c][r]:
                l1 = data[c][r].split(' ')
                for x in l1:
                    x = x.split('_')
                    l.extend(x)
                l = [ch[0].upper()+ch[1:] for ch in l]
                text = '_'.join(l)
                data[c][r] = text
            else:
                l = data[c][r].strip().split('_')
                l = [ch[0].upper()+ch[1:] for ch in l]
                text = '_'.join(l)
                data[c][r] = text
    return data
```

```python
def filtrer(l2,df_concepts):
    i=0
    indexes = []
    concepts = list(df_concepts['Concept'])
    for ch in l2:
        l = ch.split(' ')
        subs = l[0]
        links = l[1]
        obs = l[2]
        if subs.find('Process')!=-1 :
            if subs.find('Processes')!=-1:
                words = subs.split('_')
                words = [text for text in words if text!='Processes']
                subs = '_'.join(words)
            else:
                words = subs.split('_')
                words = [text for text in words if text!='Process']
                subs = '_'.join(words)
        elif obs.find('Process')!=-1 :
            if obs.find('Processes')!=-1:
                words = obs.split('_')
                words = [text for text in words if text!='Processes']
                obs = '_'.join(words)
            else:
                words = obs.split('_')
                words = [text for text in words if text!='Process']
                obs = '_'.join(words)
        if subs in concepts or obs in concepts:
            indexes.append(i)
        new_ch = subs+' '+links+' '+obs
        l2[i] = new_ch
        i+=1
    l2 = [l2[j] for j in range(len(l2)) if j not in indexes]
    return l2
l2 = filtrer(l2,df_concepts)
l2
```
